Since we only need the names of the CpG sites and not the methylation values for the samples to be able to retrieve the location of the CpG sites, we only need the column 'CpG sites' from the data table 'methylation_data'. This can be achieved by using the R notation 'data_table$column_name'.

As we can see from the output above, all the CpG sites in the 'CpG_sites' array are unique.

#### Retrieving the locations

The first step towards retrieving the location data of the CpG sites is to retrieve the annotation information for each CpG site on the specified Illumina Human Methylation 450k array. This can be achieved by calling the function 'getAnnotation' from the 'minfi' package. The function takes as a parameter the name of the library which in our case contains the annotation data from the Illumina Human Methylation 450k array with the library being IlluminaHumanMethylation450kanno.ilmn12.hg19 which was loaded into this notebook before. This library should be used as the methylation data retrieved from the GDC Data Portal originates from the illumina human methylation 450 platform. The 'ilmn12' within the library name refers to the Illumina 12th generation annotation for the array design and the 'hg19'  refers to the Human Genome version 19 assembly, which is the reference genome assembly used to map the probes on the Illumina 450K array to their respective genomic locations.

It is important to note here that the exact locations for the CpG sites will not be the same for every sample we retrieved as the genome between two samples will be different because of duplications and deletions. We can, however, use the reference genome 'hg19' as the reference genome is generally considered to be a good estimate for the positions of the CpG sites in the majority of individuals of that species. For that same reason, we can also use the reference genome in a later step to retrieve the lengths of the chromosomes (as they also differ slightly between the samples).

```r
# Retrieving the annotation information for the CpG sites present on the Illumina Human Methylation 450k array.
annotation_450k <- getAnnotation(IlluminaHumanMethylation450kanno.ilmn12.hg19)

# Converting the retrieved DataFrame to a data table to improve readability.
annotation_450k <- data.table::as.data.table(data.matrix(annotation_450k))

cat("The annotation information for the CpG sites present on the Illumina Human Methylation 450k array:")
annotation_450k
```

Now that we have succesfully loaded the annotation information for the CpG sites present on the Illumina Human Methylation 450k array, we can check whether indeed all the CpG sites present within the methylation data stored in the variable 'CpG_sites' is also present in the annotation data stored in the variable 'annotation_450k'. This can be achieved by applying the '%in%' operator which in the code below checks whether the 'CpG_sites' appear in the 'Name' column in the 'annotation_450k' data table. If a CpG site is not present in that column, it will be marked as 'TRUE' within the 'missing_sites' string of characters. 

```r
# Checking whether the 'CpG_sites' appear in the 'Name' column in the 'annotation_450k' data table.
missing_sites <- !CpG_sites %in% annotation_450k$Name

# Retrieving the CpG sites which are not present in the 'Name' column in the 'annotation_450k' data table.
CpG_sites_missing = CpG_sites[missing_sites]

# Displaying the missing sites.
if (length(CpG_sites_missing) == 0) {
    cat("All the CpG sites are present in the annotation data.")
} else {
    cat("The following CpG sites are not present in the annotation data:\n")
    cat(CpG_sites_missing)
    cat("\n\nThe number of CpG sites that are not present in the annotation data:\n")
    cat(length(CpG_sites_missing))
}
```

As we can see from the output above, all the CpG sites from the methylation data are present in the annotation data meaning that at least some information for these sites has been stored. Before we can use this annotation data to retrieve the locations of the CpG sites, we also want to check whether there are any CpG sites which, although they are present in the annotation data, contain some missing data or unexpected data for the columns (information) we want to retrieve: the chromosome, position, and strand. All the to be deleted CpG sites can then be added to the array 'CpG_sites_to_delete' such that we can later delete all of them at once.

##### The chromosome

For the chromosome, we can first of all check whether there are any CpG sites for which there is no information about the chromosome present within the 'annotation_450k' data table. This can be achieved by finding the subset of rows (CpG sites) within the 'annotation_450k' data table where the CpG site is of course present in the 'CpG_sites' character array and has a missing value for that column. The function 'subset() can help to find these subsets.

```r
# Retrieving the CpG sites for which the column 'chr' representing the chromosome has a missing value.
missing_sites_chromosome <- subset(annotation_450k, Name %in% CpG_sites & is.na(chr))

# Checking whether the number of CpG sites in the 'missing_sites_chromosome' is equal to 0.
if (nrow(missing_sites_chromosome) == 0) {
    cat("None of the CpG sites present in the 'CpG_sites' character array have a missing value for the 'chr' column.\n")
} else {
    cat("The following CpG sites that are present in the 'CpG_sites' character array have a missing value for the 'chr' column:\n")
    print(missing_sites_chromosome)
    cat("\n\nThe number of CpG sites that have a missing value for the 'chr' column.:\n")
    cat(nrow(missing_sites_chromosome))
}
```

As we can see from the output above, none of the CpG sites present in the 'CpG_sites' character array have a missing value for the 'chr' column of the 'annotation_450k' data table. Next, we need to verify whether all the values in the 'chr' column correspond to values that we would expect to see. In this case, we expect the values to be in the set chr1 to chr22 (as we also want to exclude the sex chromosomes).

```r
# Checking whether all the values in the 'chr' column correspond to values that we would expect to see. In this case, we 
# expect the values to be in the set chr1 to chr22 (as we also want to exclude the sex chromosomes).
expected_chromosomes <- c(1:22)
sites_unexpected_chromosome <- subset(annotation_450k, Name %in% CpG_sites & !chr %in% paste0("chr", expected_chromosomes))

# Checking whether the number of CpG sites in the 'sites_unexpected_chromosome' is equal to 0.
if (nrow(sites_unexpected_chromosome) == 0) {
    cat("None of the CpG sites present in the 'CpG_sites' character array have an unexpected value for the 'chr' column.\n")
} else {
    cat("The following CpG sites that are present in the 'CpG_sites' character array have an unexpected value for the 'chr' column:\n")
    print(sites_unexpected_chromosome)
    cat("\n\nThe number of CpG sites that have an unexpected value for the 'chr' column.:\n")
    cat(nrow(sites_unexpected_chromosome))
}
```

As we can see from the output above, there are 10,122 CpG sites present in the 'CpG_sites' character array which have an unexpected value for the 'chr' column of the 'annotation_450k' data table. This number is so high as most of them are located on the X and Y chromosomes and on the mitochondrial DNA which contains a single circular chromosome. Since we do not want to investigate the CpG sites located on the X and Y chromosome and on the mitochondrial DNA, we can add them to the 'CpG_sites_to_delete' array such that these are deleted later as well.

```r
# Adding the 'sites_unexpected_chromosome' to the 'CpG_sites_to_delete' array. 
CpG_sites_to_delete <- c(sites_unexpected_chromosome$Name)
```

```r
# Checking whether the 'CpG_sites' appear in the 'CpG_sites_to_delete' array.
missing_CpG_sites <- CpG_sites %in% CpG_sites_to_delete

# Retrieving the CpG sites which are present in the 'CpG_sites_to_delete' array.
CpG_sites_missing = CpG_sites[missing_CpG_sites]

# Displaying the missing genes.
if (length(CpG_sites_missing) == 0) {
    cat("All the CpG sites present in the 'CpG_sites_to_delete' array have been removed.")
} else {
    cat("The following CpG sites present in the 'CpG_sites_to_delete' array have not been removed:\n")
    cat(CpG_sites_missing)
    cat("\n\nThe number of CpG sites present in the 'CpG_sites_to_delete' array that have not been removed:\n")
    cat(length(CpG_sites_missing))
}
```

We can now use this annotation data to retrieve the locations of the CpG sites. Before we can use this annotation data to retrieve the locations of the CpG sites, we should first create a data table to store the location information for each of the CpG sites present in the methylation data. This can be achieved by calling the function 'data.table()' present in the data.table library. Within this function call, we can set the column 'CpG_site' to contain the CpG sites present in the character array 'CpG_sites'.

```r
# Creating a data table to store the location information for each of the CpG sites present in the methylation data and 
# adding all these CpG sites to the column 'CpG_site'.
CpG_sites_location_data <- data.table::data.table(CpG_site = CpG_sites, chromosome = character(), position = numeric(), strand = character())

cat("The 'CpG_sites_location_data' data table:")
CpG_sites_location_data
```

To verify whether all the CpG sites from the 'CpG_sites' character array have been added to the 'CpG_sites_location_data' data table, we can check whether the number of rows corresponds to the number of elements in the 'CpG_sites' character array.

```r
# Checking whether all the CpG sites from the 'CpG_sites' character array have been added to the 'CpG_sites_location_data' 
# data table.
if (nrow(CpG_sites_location_data) == length(CpG_sites)) {
    cat("All of the CpG sites from the 'CpG_sites' character array have been added.")
} else {
    cat("Not all of the CpG sites from the 'CpG_sites' character array have been added.")
}
```

Now, we can add the location data for each of the CpG sites by retrieving the data from the 'annotation_450k' data table and inserting it into the 'CpG_sites_location_data' data table.

Since we do not want to retrieve two records for a single CpG site, we need to check whether all the CpG sites in the 'annotation_450k' data table are unique. This can be achieved by calling the function 'unique()' with as argument the column of the CpG sites.

```r
# Checking whether all the CpG sites in the 'annotation_450k' data table are unique.
unique_cpg_sites <- unique(annotation_450k$Name)

if (length(unique_cpg_sites) == nrow(annotation_450k)) {
  cat("All CpG sites in the 'annotation_450k' data table are unique.")
} else {
  cat("There are non-unique CpG sites in the 'annotation_450k' data table.")
}
```

Now that we have verified that all CpG sites in the 'annotation_450k' data table are unique, we can safely retrieve the records for a single CpG site. This can be achieved by first calling the function 'match()' which retrieves the matching rows based on the sites present in the 'CpG_site' column from the 'CpG_sites_location_data' data table and the 'Name' column from the 'annotation_450k' data table. From these matching rows, the chromosome, the position, and the strand can be retrieved to fill in the missing values in the 'CpG_sites_location_data' data table.

```r
# Matching the CpG sites in the 'CpG_sites_location_data' data table to the 'annotation_450k' data table.
matching_rows <- match(CpG_sites_location_data$CpG_site, annotation_450k$Name)

# Filling in the missing values in the 'CpG_sites_location_data' data table with the chromosome, position, and strand from 
# the 'annotation_450k' data table.
CpG_sites_location_data$chromosome <- annotation_450k$chr[matching_rows]
CpG_sites_location_data$position <- as.numeric(annotation_450k$pos[matching_rows])
CpG_sites_location_data$strand <- annotation_450k$strand[matching_rows]
```

To increase the readability of the data table and increase the structure we can sort the table based on the 'chromosome' column followed by the 'position' column. To define the order in which the chromosomes should be sorted, the function 'factor' can be used first. To actually order the data table, we can call the function 'order()'.

```r
# Defining the order in which the chromosomes should be sorted by calling the function 'factor()'.
CpG_sites_location_data$chromosome <- factor(CpG_sites_location_data$chromosome, levels = c("chr1", "chr2", "chr3", "chr4", "chr5", "chr6", "chr7", "chr8", "chr9", "chr10", "chr11", "chr12", "chr13", "chr14", "chr15", "chr16", "chr17", "chr18", "chr19", "chr20", "chr21", "chr22"))

# Sorting the table based on the chromosome and the position to increase the structure by calling the function 'order()'.
CpG_sites_location_data_sorted = CpG_sites_location_data[order(CpG_sites_location_data$chromosome, position)]

cat("The resulting 'CpG_sites_location_data_sorted' data table:")
CpG_sites_location_data_sorted
```

Now the location data of the CpG sites has been retrieved, we can store the resulting data table 'CpG_sites_location_data_sorted' featuring this information. This can be achieved by calling the function 'fwrite()' from the 'data.table' library which takes as arguments the data table to be stored and the location where the data table should be stored.

```r
# The path where the 'CpG_sites_location_data_sorted' data table should be stored.
path <- file.path(data_directory_location_files, "CpG_sites_location_data.csv")

# If the 'path' defined above does already point to a file, the execution of this code block is terminated and a message is 
# displayed informing that the file already exists.
if (file.exists(path)) {
    cat(paste("There is already a file present at the path: ", path))
} else {
    # Writing the data table to a CSV file.
    data.table::fwrite(CpG_sites_location_data_sorted, path)
    cat(paste("The file has been created at the path: ", path))
}
```

## Retrieving the Lengths of the Chromosomes

Since some of the CpG sites that lie on the same chromosome are located on a different strand, we might need the lengths of the chromosomes to be able to calculate the distance between such two CpG sites. Even though the positions present in the 'CpG_sites_location_data_sorted' all start from the same side of the chromosome no matter what the strand is and thus a simple absolute subtraction returns the distance, it might still be handy to retrieve the lengths of the chromosomes right now as they might be useful later on.

As explained before, it is important to note here that the distance between CpG sites will not be the same for every sample we retrieved as the genome between two samples will be different because of duplications and deletions. We can, however, use the reference genome 'hg19' (human genome version h19), which was also the reference genome used by the IlluminaHumanMethylation450kanno.ilmn12.hg19 annotation package to retrieve the locations of the CpG sites, as the reference genome is generally considered to be a good estimate for the location data of the CpG sites as well as the lengths of the chromosomes in the majority of individuals of that species.

Before we retrieve the lengths of the chromosomes in that reference genome, we should first create a data table to store the lengths. This can be achieved by calling the function 'data.table()' present in the data.table library. Within this function call, we can set the column 'chromosome' to contain the chromosomes present in the 'chromosome' column of the 'CpG_sites_location_data_sorted' data table by calling the function 'levels()' on that column.

```r
# Creating a data table to store the lengths of each of the chromosomes present in the 'CpG_sites_location_data_sorted' data 
# table and adding all the chromosomes present in the column 'chromosome' of that data table.
chromosomes_length_data <- data.table::data.table(chromosome = levels(CpG_sites_location_data_sorted$chromosome), length = numeric())

cat("The 'chromosomes_length_data' data table:")
chromosomes_length_data
```

Now to retrieve the lengths of each of these chromosomes, we can call the function 'seqlengths()' with as argument the 'Hsapiens' library loaded into this notebook before. Since in addition to the lengths of the regular chromosomes also the lengths of some haplotypes and random contigs are retrieved, we can restrict the function to only retrieve the lengths for the chromosomes present in the 'chromosomes_length_data' data table.

```r
# Retrieving the lengths of the chromosomes present in the 'chromosomes_length_data' data table by calling the function 
# 'seqlengths()' with as argument the 'Hsapiens' library.
chromosome_lengths <- seqlengths(Hsapiens)[1:22]

cat("The lengths of the chromosomes present in the 'chromosomes_length_data' data table:")
chromosome_lengths
```

We can add these lengths by looping over the 'chromosome_lengths' array and calling the function 'seq_along()'.

```r
# Looping over the lengths present in the 'chromosome_lengths' array.
for (i in seq_along(chromosome_lengths)) {
    chromosomes_length_data$length[i] <- as.numeric(chromosome_lengths[i])
}

cat("The 'chromosomes_length_data' data table:")
chromosomes_length_data
```

Now as a final check we can investigate whether the length of each chromosomes is higher than the maximum position of the CpG sites present in the 'CpG_sites_location_data' that are located on that chromosome. 

The first step is to retrieve the maximum positions of the CpG sites for each of the chromosomes. This can be achieved by using the pipe operator '%>%' which takes the output of the code on the left and passes it as the first argument to the code on the right. The function 'groupby()' is called to group the data by chromosome. The function 'summarise()' is then called for each of these groups to calculate the maximum position within the 'position' column. 

```r
# Retrieving the maximum positions of the CpG sites for each of the chromosomes.
max_positions <- CpG_sites_location_data %>% group_by(chromosome) %>% summarise(max_position = max(position))

cat("The maximum positions of the CpG sites for each of the chromosomes:")
max_positions
```

The next step then is to join the 'max_positions' and 'chromosomes_length_data' based on the 'chromosome' column which in the last step allows us to more easily compare the values. This can be achieved by calling the function 'left_join()'.

```r
# Joining the 'max_positions' and 'chromosomes_length_data' based on the 'chromosome' column.
chromosome_info <- chromosomes_length_data %>% left_join(max_positions, by = "chromosome") %>% drop_na()

cat("All the necessary information about the chromosomes to check whether the length of each chromosomes is higher than the maximum position of the CpG sites present in the 'CpG_sites_location_data' that are located on that chromosome:")
chromosome_info
```

As the final step, we can now actually check whether the length of each chromosomes is higher than the maximum position of the CpG sites present in the 'CpG_sites_location_data' that are located on that chromosome. This can be achieved by calling the function 'mutate()' which adds another column called 'is_valid' containing whether the length for that chromosome is indeed larger than the maximum position.

```r
# Retrieving whether the length for each chromosome is larger than its maximum position.
chromosome_info <- chromosome_info %>% mutate(is_valid = length > max_position)

cat("The chromosome information containing whether the length for each chromosome is larger than its maximum position:")
chromosome_info
```

Since for each chromosome its length is larger than its maximum position, the chromosome length data present in the 'chromosomes_length_data' data table is valid.
