## Introduction

Within this notebook, the normalized gene expression data files are checked to see whether the samples for each of the gene expression data files are indeed quantile normalized. This can be achieved by for each of the normalized gene expression data files plotting a box and whisker plot for each of the samples their distribution within the same plot and then seeing whether these boxes align. If they all do align, this means that their distributions are the same meaning the quantile normalization process has been performed correctly (in addition to already knowing that the samples in the test set were not used to influence the quantile normalization process for the samples in the training set meaning there is little to no information leakage between the sets).

## Checking Whether the Gene Expression Data Files are Correctly Quantile Normalized

Within this section, we can check whether the normalized gene expression data files from the directory 'data_directory_final_datasets' are correctly quantile normalized. For each data file a different section is defined below where we check this. Within each of these sections, the corresponding gene expression data is loaded into the notebook by calling the function 'pd.read_csv()' and a box and whisker plot for each of the samples their distribution is plotted to see whether the boxes (their distributions) align by calling the function 'sns.boxplot()'.

#### Checking whether the 'gene_expression_data_normalized_split1.csv' file is correctly quantile normalized

```python
# Loading the 'gene_expression_data_normalized_split1.csv' file into this notebook by calling the function 'pd.read_csv()'.
gene_expression_data_normalized_split1 = pd.read_csv(data_directory_final_datasets + "/" + "gene_expression_data_normalized_split1.csv")

# Transposing the DataFrame by calling the function 'transpose()' as the function 'sns.boxplot()' requires the samples to be
# represented by the columns.
gene_expression_data_normalized_split1.set_index(gene_expression_data_normalized_split1.columns[0], inplace=True)
gene_expression_data_normalized_split1 = gene_expression_data_normalized_split1.transpose()

print("The 'gene_expression_data_normalized_split1' DataFrame containing the data from the 'gene_expression_data_normalized_split1.csv' file:")
gene_expression_data_normalized_split1
```

```python
plt.figure(figsize=(20, 12))

# Creating a boxplot for every column (sample) in the 'gene_expression_data_normalized_split1' DataFrame, plotting them on 
# the same axis, without showing the outliers.
ax = sns.boxplot(data=gene_expression_data_normalized_split1, showfliers=False)
ax.set_xticklabels(ax.get_xticklabels(), rotation='vertical')

# Adding the legend, a title and the labels to the plot.
plt.title("The Distributions of the Samples for the 'gene_expression_data_normalized_split1' DataFrame")
plt.xlabel("Sample")
plt.ylabel("Gene Expression Value")

# Show the plot
plt.show()
```

As we can see from the plot above, the distributions of all the samples is the same indicating that the quantile normalization has been correctly applied (in addition to already knowing that the samples in the test set were not used to influence the quantile normalization process for the samples in the training set meaning there is little to no information leakage between the sets).
